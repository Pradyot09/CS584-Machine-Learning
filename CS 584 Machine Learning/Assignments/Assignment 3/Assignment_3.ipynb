```python
def EntropyIntervalSplit (
    inData,          # input data frame (predictor in column 0 and target in column 1)
    split):          # split value
    index= []

    dataTable = inData

    crossTable = pd.crosstab(index = (inData.iloc[:,0]).isin(split), columns = dataTable.iloc[:,1], margins = True, dropna = True)   
    nRows = crossTable.shape[0]
    nColumns = crossTable.shape[1]

    tableEntropy = 0
    for iRow in range(nRows-1):
        rowEntropy = 0
        for iColumn in range(nColumns):
            proportion = crossTable.iloc[iRow,iColumn] / crossTable.iloc[iRow,(nColumns-1)]
            if (proportion > 0):
                rowEntropy -= proportion * np.log2(proportion)
        print('Row = ', iRow, 'Entropy =', rowEntropy)
        index.append(rowEntropy)
        print(' ')
        tableEntropy += rowEntropy *  crossTable.iloc[iRow,(nColumns-1)]
    tableEntropy = tableEntropy /  crossTable.iloc[(nRows-1),(nColumns-1)]

    return(tableEntropy, index)
```

```python
print("1(c)")
import itertools as itr
splitcounter=0
decisionTreeCard = pd.DataFrame(columns = ['SplitIndex','Left_Child', 'Right_Child', 'SplitEntropy','Counter'])
for i in range(1, categoryOfCards):
    splitcounter+=1
    print('Number of Elements in the Split Set = ', i)
    totalCombinations = itr.combinations(uniqueCardValues, i)
    for combinations in list(totalCombinations):
        restOfCards = uniqueCardValues.difference(combinations)
        print('Selected Cards: ', combinations)
        print('Rest of the cards: ', restOfCards)
        SplitEntropyCard, SplitIndexC= EntropyIntervalSplit(inDataC, combinations)
        print('Split Entropy: ', SplitEntropyCard)
        decisionTreeCard = decisionTreeCard.append(pd.DataFrame([[SplitIndexC,sorted(combinations), sorted(restOfCards), SplitEntropyCard,splitcounter]], 
                                              columns = ['SplitIndex','Left_Child', 'Right_Child', 'SplitEntropy','Counter']))
        

print('Decision tree of Card: \n', decisionTreeCard)
```
